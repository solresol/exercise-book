Convert a category into a several columns of numbers

```python
sacramento_with_dummies = pandas.get_dummies(sacramento, columns=['dwelling_type'])
sacramento_with_dummies
```

If we attempt a linear regression using all 4 dummy variables, there
are an infinite number of options for the zero-intercept point.

sklearn chooses the lowest possible float for this: -1.4880279789385556e+18

```python
import sklearn.linear_model
ols = sklearn.linear_model.LinearRegression()
columns = ['dwelling_type_Condo', 'dwelling_type_Multi-Family',
           'dwelling_type_Residential', 'dwelling_type_Unkown']
ols.fit(sacramento_with_dummies[columns], sacramento_with_dummies.price)
zip(columns, ols.coef_), ols.intercept_
```

```python
Even ignoring one column gives us sensible, comprehensible results.
```

We can use this on the Sacramento, perhaps to get a street name or street number

```python
sacramento.street.str.get_dummies(sep=" ")
```

But that dataframe is wider than our dataset is long. Bad news!

A better approach would be to look for specific features

```python
sacramento['bad_fengshui'] = sacramento.street.str.contains('4')
```

We might need to do a regular expression looking for the last few
non-space characters

```python
sacramento['last_part_of_street'] = sacramento.street.str.extract('.* ([^ ]*)$')
```

Or a variation:

```python
sacramento['type_of_street'] = sacramento.street.str.extract('(CT|ST|DR|WAY|AVE|CIR|RD|PL|PKWY|BLVD|LN)')
```

# Examples of cut and qcut

```python
%matplotlib inline
import seaborn
```

```python
seaborn.distplot(sacramento.sq__ft)
```

```python
sacramento.sq__ft.nlargest(100).min()
```

We can split at 100 per bin from the top, but there would be
more than 100 in the bottom bin. We confirm that 200 in the bottom
bin is possible.

```python
sacramento.sq__ft.nsmallest(200).max()
```

So five bins is possible. This is what they would look like.

```python
pandas.qcut(sacramento.sq__ft,5).value_counts()
```

```python
sacramento['footage_size'] = pandas.qcut(sacramento.sq__ft,5)
```

We could do prettier naming.

```python
pandas.qcut(sacramento.sq__ft,5, labels=['Minimal', 'Small', "Medium", "Large", "Farm"]).value_counts()
```

```python
sacramento['footage_description'] = pandas.qcut(
  sacramento.sq__ft,5, labels=['Minimal', 'Small', "Medium", "Large", "Farm"]
)
```

Alternatively, we might choose equal-sized footage ranges

```python
pandas.cut(sacramento.sq__ft, 10).value_counts()
```

```python
seaborn.distplot(sacramento.sq__ft, 10)
```

```python
sacramento['footage_range'] = pandas.cut(sacramento.sq__ft, 10)
```

... and then we would convert these categories to dummy variables
and regress against the dummy variables.

This makes sense if there isn't a linear relationship between sqft and

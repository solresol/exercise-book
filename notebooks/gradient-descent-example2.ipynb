```python
import pandas
import math
import random
```

```python
a = 0.73
b = 0.18
c = 0.37
```

```python
import numpy
ks = numpy.arange(0, 8, 0.01)
```

```python
noise = [random.random() for x in ks]
```

```python
def func(t):
    return math.sin(a * t + b) + c

data = pandas.Series(data=map(func, ks), index=ks) + pandas.Series(noise, index=ks)
```

```python
%matplotlib inline
data.plot()
```

```python
csv = pandas.DataFrame({
  'seconds': ks,
  'signal': data
})
```

```python
csv
```

```python
csv.plot.scatter('seconds', 'signal')
```

```python
csv[['signal', 'seconds']].to_csv('wheel.csv')
```

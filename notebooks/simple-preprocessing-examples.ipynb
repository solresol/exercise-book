# Date columns

```python
pandas.datetime.strptime('Mar-14', '%b-%y')
```

```python
def dateparse(x):
    try:
        return pandas.datetime.strptime(x, '%d/%m/%Y')
    except ValueError:
        pass
    try:
        return pandas.datetime.strptime(x, "%b-%y")
    except ValueError:
        pass
    if "-" in x:
        (lhs, rhs) = x.split('-',1)
        return dateparse(rhs)

sydtrains = pandas.read_csv('sydtrains.csv',
                            index_col=1,
                            parse_dates=['SURVEY_DATE_USED'],
                            date_parser=dateparse
                            )
sydtrains
```

Let's check some data types:

```python
sydtrains.SURVEY_DATE_USED
```

expected int64

# Dealing with commas in numbers

European datasets often use . as a thousands separator. This
dataset uses , in several places.

```python
sydtrains = pandas.read_csv('sydtrains.csv',
                            index_col=1,
                            parse_dates=['SURVEY_DATE_USED'],
                            date_parser=dateparse,
                            thousands=","
                            )
sydtrains
```

```python
sydtrains.OUT_24_HOURS
```

Confirmed, the column is now numbers, not a mixture of numbers and strings.

# Restricting to one particular category

Perhaps we are only interested in a particular year, e.g. the last year before the Opal card was introduced.

```python
trains2014_selector = sydtrains.YEAR == 2014
```

```python
trains2014 = sydtrains[trains2014_selector].copy()
trains2014
```

We needed to say "copy()" above so that we can add extra columns.

(Otherwise we would be trying to add columns to a view over the existing
data. Instead, we want to edit a fresh dataframe).

```python
trains2014['differences'] = trains2014.IN_24_HOURS - trains2014.OUT_24_HOURS
```

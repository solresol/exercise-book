# Multi-dimensional linear regression

```python
sacramento = pandas.read_csv('Sacramentorealestatetransactions.csv',
                             index_col=0
                             )
sacramento
```

```python
sacramento.columns
```

```python
X = sacramento[['beds', 'baths', 'sq__ft', 'latitude', 'longitude']]
```

```python
y = sacramento.price
```

```python
seaborn.pairplot(sacramento)
```

```python
ols = sklearn.linear_model.LinearRegression()
ols.fit(X, y)
zip(['beds', 'baths', 'sq__ft', 'latitude', 'longitude'],ols.coef_)
```

```python
ols.score(X,y)
```

```python
t2 = sklearn.linear_model.TheilSenRegressor()
t2.fit(X,y)
zip(['beds', 'baths', 'sq__ft', 'latitude', 'longitude'],t2.coef_)
```

```python
t2.score(X,y)
```

```python
r2 = sklearn.linear_model.RANSACRegressor()
r2.fit(X,y)
zip(['beds', 'baths', 'sq__ft', 'latitude', 'longitude'],r2.estimator_.coef_)
```

```python
lasso = sklearn.linear_model.Lasso(alpha=10000)
lasso.fit(X,y)
zip(['beds', 'baths', 'sq__ft', 'latitude', 'longitude'],lasso.coef_)
```

```python
lasso.score(X,y)
```

```python
lasso_prediction = lasso.predict(X)
matplotlib.pyplot.scatter(sacramento.baths, sacramento.price)
matplotlib.pyplot.scatter(sacramento.baths, lasso_prediction, color="red")
```

```python
import scipy.stats
scipy.stats.mannwhitneyu(p9575.price, p9576.price)
```

```python
for pcode in sacramento.postcode_region.unique():
    for qcode in sacramento.postcode_region.unique():
        if pcode > qcode: continue
        p1 = sacramento[sacramento.postcode_region == pcode]
        p2 = sacramento[sacramento.postcode_region == qcode]
        print pcode, qcode, scipy.stats.mannwhitneyu(p1.price, p2.price)
```

```python
%matplotlib inline
import pandas
```

```python
sat = pandas.read_csv('sat_scores.csv', index_col=0)
```

```python
abbrevs_url = 'https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States'
abbrevs_page = pandas.read_html(abbrevs_url, header=0)
abbrevs_page[0]
```

```python
populations_url = 'https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population'
populations_page = pandas.read_html(populations_url, header=0)
populations_page[1]
```

```python
abbrev_to_pops = abbrevs_page[0].merge(populations_page[1], on="State")
abbrev_to_pops[['Abbr.','Population']]
```

```python
extended_sat = sat.merge(abbrev_to_pops[['Abbr.','Population']], 
          left_index=True, 
          right_on="Abbr.")
extended_sat
```

```python
extended_sat.shape
```

```python
import seaborn
seaborn.lmplot('Math', 'Verbal', data=extended_sat)
```

```python
seaborn.lmplot('Rate', 'Verbal', data=extended_sat)
```

```python
highest_bad_rate = extended_sat.Rate.nsmallest(10).max()
highest_bad_rate
```

```python
lowest_good_rate = extended_sat.Rate.nlargest(10).min()
lowest_good_rate
```

```python
seaborn.distplot(extended_sat.Math)
```

```python
seaborn.distplot(extended_sat.Verbal)
```

```python
low_rate_sats = extended_sat[extended_sat.Rate <= highest_bad_rate]
high_rate_sats = extended_sat[extended_sat.Rate >= lowest_good_rate]
```

```python
import statsmodels.stats.weightstats
ztest = statsmodels.stats.weightstats.ztest

ztest(low_rate_sats.Math, value=extended_sat.Math.mean())
```

```python
ttest = statsmodels.stats.weightstats.ttest_ind
ttest(low_rate_sats.Math, high_rate_sats.Math, value=extended_sat.Math.mean())
```

```python
import scipy.stats
u_test = scipy.stats.mannwhitneyu
u_test(low_rate_sats.Math, high_rate_sats.Math)
```

```python
extended_sat['Population_Bin'] = pandas.qcut(extended_sat.Population,
                                             q=5,
                                             labels=['Tiny','Small',
                                                     "Medium","Big","Huge"])
```

```python
(extended_sat.Population_Bin).value_counts()
```

```python
extended_sat[extended_sat.Population_Bin == 'Huge'][["Abbr.", 'Population']]
```

```python
extended_sat[extended_sat.Population_Bin == 'Tiny'][["Abbr.", 'Population']]
```

```python
ttest(extended_sat[extended_sat.Population_Bin == 'Huge'].Math,
      extended_sat[extended_sat.Population_Bin == 'Tiny'].Math
      )
```

```python
u_test(extended_sat[extended_sat.Population_Bin == 'Huge'].Math,
      extended_sat[extended_sat.Population_Bin == 'Tiny'].Math
      )
```

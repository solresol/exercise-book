Seaborn
=======

Seaborn is a plotting library for making really pretty plots.

```python
%matplotlib inline
import pandas
```

```python
import seaborn
```

```python
def dateparse(x):
    try:
        return pandas.datetime.strptime(x, '%d/%m/%Y')
    except ValueError:
        pass
    try:
        return pandas.datetime.strptime(x, "%b-%y")
    except ValueError:
        pass
    if "-" in x:
        (lhs, rhs) = x.split('-',1)
        return dateparse(rhs)
```

```python
all_trains = pandas.read_csv('sydtrains.csv',
                            index_col=1,
                            parse_dates=['SURVEY_DATE_USED'],
                            date_parser=dateparse,
                            thousands=","
                            )
all_trains
trains = all_trains[all_trains.YEAR == 2014].copy()
trains
```

```python
seaborn.distplot(trains.IN_0600_0930)
```

```python
with seaborn.axes_style("white"):
    seaborn.distplot(trains.IN_0600_0930)
```

```python
import matplotlib.pyplot
with seaborn.axes_style("whitegrid"):
    fig, axes = matplotlib.pyplot.subplots(1,2, figsize=(16,5))
    seaborn.distplot(trains.IN_0600_0930, ax=axes[0])
    seaborn.distplot(trains.OUT_0600_0930, ax=axes[1])
fig, axes
```

# Bivariate analysis


```python
seaborn.jointplot(x="IN_0600_0930", y="OUT_0600_0930" ,data=trains)
```

```python
seaborn.jointplot(x="IN_0600_0930", y="OUT_0600_0930" ,data=trains, color="red")
```

```python
trains.LINE.unique()
```

```python
def line_colouring(x):
    if x == 'CBD': return 'red'
    if x == 'Western': return 'orange'
    if x == 'North Shore': return 'blue'
    if x == 'Eastern Suburbs': return 'gold'
    return 'green'
train_colours = trains.LINE.map(line_colouring)

matplotlib.pyplot.scatter(trains.IN_0600_0930,
                          trains.OUT_0600_0930,
                          color=train_colours)
```

```python
seaborn.jointplot(x="IN_0600_0930", y="OUT_0600_0930", 
                  data=trains[(trains.IN_0600_0930 < 100) &
                              (trains.OUT_0600_0930 < 100)],
                  kind="kde")
```

```python
trains[(trains.IN_0600_0930 < 10) & (trains.OUT_0600_0930 < 10)].LINE
```

```python
seaborn.stripplot(x="LINE", y="IN_0600_0930", data=trains, jitter=True)
```

```python
seaborn.lmplot(x="IN_0600_0930", y="OUT_0600_0930", data=trains[trains.OUT_0600_0930 < 5000])
```

```python
seaborn.residplot(x="IN_0600_0930", y="OUT_0600_0930", data=trains[trains.OUT_0600_0930 < 5000])
```

```python
seaborn.lmplot(x="IN_0600_0930", y="OUT_0600_0930", data=trains, lowess=True)
```

# Diabetes

```python
import sklearn.datasets
diabetes = sklearn.datasets.load_diabetes()
```

```python
diabetes.data.shape
```

```python
type(diabetes.data)
```

```python
diabetes.data[:,0].shape
```

```python
seaborn.distplot(diabetes.data[:,0])
```

```python
diabetes_df = pandas.DataFrame(
 data = {
    'Age': diabetes.data[:,0],
    'Sex': diabetes.data[:,1],
    'BMI': diabetes.data[:,2],
    'BloodPressure': diabetes.data[:,3],
    'B1': diabetes.data[:,4],
    'B2': diabetes.data[:,5],
    'B3': diabetes.data[:,6],
    'B4': diabetes.data[:,7],
    'B5': diabetes.data[:,8],
    'B6': diabetes.data[:,9],
    'Progression': diabetes.target
  }
)
diabetes_df
```

```python
seaborn.jointplot(x="Age", y="Progression", data=diabetes_df)
```

```python
seaborn.jointplot(x="Age", y="Progression", kind="hex", data=diabetes_df)
```

```python
seaborn.jointplot(x="B1", y="Progression", kind="hex", data=diabetes_df, bins=5)
```

```python
seaborn.jointplot(x="B1", y="Progression", kind="kde", data=diabetes_df, bins=5)
```

```python
variables = [x for x in diabetes_df.columns if x != "Progression"]
(fig, axes) = matplotlib.pyplot.subplots(5,2, sharey=True, figsize=(8,12))
for (v,a) in zip(variables,axes.flatten()):
    print v,a
    seaborn.jointplot(x=v, y="Progression", kind="kde", data=diabetes_df, ax=a)
```

```python
seaborn.jointplot(x="B5", y="Progression", data=diabetes_df)
```

```python
seaborn.jointplot(x="B5", y="Progression", data=diabetes_df, kind="reg")
```

```python
seaborn.pairplot(data=diabetes_df, hue="Sex")
```

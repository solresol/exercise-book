Or, an example from a local file

```python
import json
```

```python
yelp_data = open('yelp.json')
```

```python
yelp_data_as_a_list = yelp_data.readlines()
```

```python
len(yelp_data_as_a_list)
```

```python
json.loads(yelp_data_as_a_list[0])
```

```python
stars = []
mentions_vegetables = []
for review in yelp_data_as_a_list:
    parsed = json.loads(review)
    stars.append(parsed['stars'])
    mentions_vegetables.append('vegetable' in parsed['text'].lower())
zip(stars, mentions_vegetables)
```

```python
import pandas
df = pandas.DataFrame(
   data={
     'Stars' : stars,
     'Mentions_vegetables': mentions_vegetables
})
df
```

```python
%matplotlib inline
import seaborn
```

```python
seaborn.distplot(df[df.Mentions_vegetables].Stars)
seaborn.distplot(df[df.Mentions_vegetables==False].Stars)
```

```python
seaborn.violinplot(y="Stars", x="Mentions_vegetables", data=df, split=True)
```

```python
seaborn.boxplot(y="Stars", x="Mentions_vegetables", data=df)
```

```python
df[df.Mentions_vegetables].Stars.mean()
```

```python
df[df.Mentions_vegetables == False].Stars.mean()
```

```python
df[(df.Stars == 5) & (df.Mentions_vegetables)].count()
```

```python
df[(df.Stars == 5) & (df.Mentions_vegetables==False)].count()
```

Alternate JSON api... more object-oriented -- json.loads replaced
by a method

```python
dec = json.JSONDecoder()
stars = []
mentions_vegetables = []
for review in yelp_data_as_a_list:
    parsed = dec.decode(review)
    stars.append(parsed['stars'])
    mentions_vegetables.append('vegetable' in parsed['text'].lower())
zip(stars, mentions_vegetables)
```

```python
import pandas
billboard = pandas.read_csv('billboard.csv',
                            parse_dates=['date.entered', 'date.peaked'],
                            encoding='ISO-8859-1')
```

```python
billboard['peak_duration'] = billboard['date.peaked'] - billboard['date.entered']
billboard.peak_duration
```

```python
new_column_names = []
for column in billboard.columns:
    if column.endswith('week') and len(column)==9:
        good_name = 'x0' + column[1:]
        print "Bad column name",column,"should be",good_name
        new_column_names.append(good_name)
    else:
        new_column_names.append(column)
new_column_names
```

```python
billboard.columns = new_column_names
```

```python
billboard.mean()
```

```python
billboard['x66th.week'].isnull().all()
```

```python
billboard.median()
```

```python
billboard.std()
```

```python
import numpy
columns_that_are_numbers = billboard.columns[
     (billboard.dtypes == numpy.int64) | (billboard.dtypes == numpy.float64)
]
columns_that_are_numbers
```

```python
billboard.max()[columns_that_are_numbers] - billboard.min()[columns_that_are_numbers]
```

```python
stats = pandas.DataFrame(
 {
  'mean': billboard.mean(),
  'median': billboard.median(),
  'stddev': billboard.std(),
  'range': (
    billboard.max()[columns_that_are_numbers] -
       billboard.min()[columns_that_are_numbers]),
 }
)
stats
```

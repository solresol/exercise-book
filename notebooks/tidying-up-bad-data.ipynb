# Examples with bad data

```python
url = 'https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population'
dataframes = pandas.read_html(url, header=0)
```

```python
dataframes[0]
```

```python
dataframes[1].get_dtype_counts()
```

```python
dataframes[1].columns
```

```python
dataframes[1].dtypes
```

```python
dataframes[2].get_dtype_counts()
```

```python
dataframes[2].dtypes
```

```python
dataframes[2].isnull()
```

If we want to fill in null values with a mean, we can.
Don't forget in-place though.

```python
dataframes[2]['Growth Rate (Percent)'].fillna(
 dataframes[2]['Growth Rate (Percent)'].mean(),
 inplace=True
)
```

# Load the Sydney climate data

```python
weather = pandas.read_csv('sydney_climate.csv', parse_dates='Year_Month_Day')
weather
```

```python
weather.get_dtype_counts()
```

```python
weather.dtypes
```

```python
%matplotlib inline
```

```python
weather.MANGROVE_MOUNTAIN.plot()
```

```python
weather.MANGROVE_MOUNTAIN.mean()
```

```python
pandas.rolling_mean(weather.MANGROVE_MOUNTAIN, 3)
```

```python
weather.MANGROVE_MOUNTAIN.fillna(method='pad')
```

```python
weather.TERREY_HILLS.fillna(method='pad', limit=1).fillna(method='bfill', limit=1)
```

```python
weather.dropna(inplace=True)
weather
```

```python
%matplotlib inline
import pandas
```

```python
weather = pandas.read_csv('sydney_climate.csv')
```

```python
weather
```

```python
weather.SYDNEY
```

```python
weather[['SYDNEY', 'SYDNEY_AIRPORT', "MANGROVE_MOUNTAIN"]].plot()
```

```python
weather.dropna(inplace=True)
```

```python
import seaborn
seaborn.pairplot(weather[['SYDNEY', 'SYDNEY_AIRPORT', "MANGROVE_MOUNTAIN"]])
#seaborn.pairplot(weather)
```

```python
import sklearn.linear_model
lr = sklearn.linear_model.LinearRegression()

non_sydney_columns = [x for x in weather.columns if x not in ['SYDNEY', 'Year_Month_Day']]

lr.fit(weather[non_sydney_columns], weather.SYDNEY)
predictions = lr.predict(weather[non_sydney_columns])
import matplotlib.pyplot
matplotlib.pyplot.scatter(weather.SYDNEY, predictions)
```

```python
lr.score(weather[non_sydney_columns], weather.SYDNEY)
```

```python
zip(non_sydney_columns, lr.coef_)
```

```python
lasso = sklearn.linear_model.Lasso(alpha=5)
lasso.fit(weather[non_sydney_columns], weather.SYDNEY)
print lasso.score(weather[non_sydney_columns], weather.SYDNEY)
print lasso.intercept_
zip(non_sydney_columns, lasso.coef_)
```

```python
ridge = sklearn.linear_model.Ridge(alpha=100)
ridge.fit(weather[non_sydney_columns], weather.SYDNEY)
print ridge.score(weather[non_sydney_columns], weather.SYDNEY)
print ridge.intercept_
zip(non_sydney_columns, ridge.coef_)
```

```python
elastic = sklearn.linear_model.ElasticNetCV()
elastic.fit(weather[non_sydney_columns], weather.SYDNEY)
print elastic.score(weather[non_sydney_columns], weather.SYDNEY)
print elastic.intercept_
zip(non_sydney_columns, elastic.coef_)
```

# Scaling

Of course, scaling matters, because the range of these temperatures is not exactly the same.

```python
weather[non_sydney_columns].mean()
```

```python
weather[non_sydney_columns].std()
```

```python
import sklearn.preprocessing
robust = sklearn.preprocessing.RobustScaler()
scaled_weather = robust.fit_transform(weather[non_sydney_columns])
scaled_weather.mean(), scaled_weather.std()
```

```python
lasso = sklearn.linear_model.Lasso(alpha=1)
lasso.fit(scaled_weather, weather.SYDNEY)
print lasso.score(scaled_weather, weather.SYDNEY)
print lasso.intercept_
zip(non_sydney_columns, lasso.coef_)
```

```python
y_scaler = sklearn.preprocessing.RobustScaler()
sydney_transformed = y_scaler.fit_transform(weather[['SYDNEY']])
```

```python
lasso = sklearn.linear_model.Lasso(alpha=0.1)
lasso.fit(scaled_weather, sydney_transformed)
print lasso.score(scaled_weather, sydney_transformed)
print lasso.intercept_
zip(non_sydney_columns, lasso.coef_)
```

```python
elastic = sklearn.linear_model.ElasticNetCV()
elastic.fit(scaled_weather, sydney_transformed)
print elastic.score(scaled_weather, sydney_transformed)
print elastic.intercept_
zip(non_sydney_columns, elastic.coef_)
```

```python
elastic.predict(scaled_weather)[0]
```

```python
y_scaler.inverse_transform([[0.02422972535434223]])
```

```python
transformed_space_predictions = elastic.predict(scaled_weather)
reality_predictions = y_scaler.inverse_transform(transformed_space_predictions)
matplotlib.pyplot.scatter(reality_predictions, weather.SYDNEY)
```
